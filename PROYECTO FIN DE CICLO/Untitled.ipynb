```python
sexo = ['Masculino', 'Femenino']
```

```python
zona = ['ZONA 1', 'ZONA 2', 'ZONA 3', 'ZONA 4', 'ZONA 5', 'ZONA 6', 'ZONA 7','ZONA 8','ZONA 9', 'NO DETERMINADA']
```

```python
provincia = ['AZUAY', 'BOLIVAR', 'CANAR', 'CARCHI','CHIMBORAZO', 
             'COTOPAXI', 'EL ORO', 'ESMERALDAS', 'GUAYAS', 
             'IMBABURA', 'LOJA', 'LOS RIOS', 'MANABI', 
             'MORONA SANTIAGO', 'NAPO','ORELLANA', 'PASTAZA', 
             'PICHINCHA', 'SANTA ELENA', 'SANTO DOMINGO DE LOS TSACHILAS',
             'SUCUMBIOS', 'TUNGURAHUA', 'ZAMORA CHINCHIPE']
```

```python
color =magma(10)
source = ColumnDataSource(data=dict(zona=zona, edadPromedioPorZona=edadPromedioPorZona, color=color))
output_file("promediosZona.html")
p = figure(x_range=zona, plot_height=500,plot_width=1000, title="Promedio de Edad de Levantamiento de Cadaveres por Zonas Delimitadas en el mes de Julio de 2014",
           toolbar_location=None, tools="")
p.vbar(x='zona', top='edadPromedioPorZona', width=0.8, color='color', legend = '', source = source)
p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
color =inferno(23)
source = ColumnDataSource(data=dict(provincia=provincia, edadPromedioPorProvincia=edadPromedioPorProvincia, color=color))
output_file("promediosProvincia.html")
q = figure(x_range=provincia, plot_height=500,plot_width=1000, title="Promedio de Edad de Levantamiento de Cadaveres por Provincias en el mes de Julio de 2014",
           toolbar_location=None, tools="")
q.vbar(x='provincia', top='edadPromedioPorProvincia', width=0.8, color='color', legend = '', source = source)
q.xgrid.grid_line_color = None
q.xaxis.major_label_orientation = math.pi/2
q.y_range.start = 0

show(q)
```

```python
color = Spectral6[4:6]
source = ColumnDataSource(data=dict(sexo=sexo, edadPromedioPorSexo=edadPromedioPorSexo, color=color))
output_file("promediosSexo.html")
r = figure(x_range=sexo, plot_height=500,plot_width=700, title="Promedio de Edad de Levantamiento de Cadaveres por Sexo en el mes de Julio de 2014",
           toolbar_location=None, tools="")
r.vbar(x='sexo', top='edadPromedioPorSexo', width=0.5, color='color', legend = '', source = source)
r.xgrid.grid_line_color = None
r.y_range.start = 0

show(r)
```
